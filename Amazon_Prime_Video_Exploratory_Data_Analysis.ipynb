## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


# 1. Duplicate values check & remove

print("Duplicates in Titles:", titles.duplicated().sum())
print("Duplicates in Credits:", credits.duplicated().sum())

titles.drop_duplicates(inplace=True)
credits.drop_duplicates(inplace=True)

# 2. Missing values check

print("\nMissing values in Titles:")
print(titles.isnull().sum())
print("\nMissing values in Credits:")
print(credits.isnull().sum())


# 3. Handle missing values

titles['age_certification'] = titles['age_certification'].fillna('Unknown')

# Fill numeric columns with 0 where missing makes sense
titles['runtime'] = titles['runtime'].fillna(0)
titles['seasons'] = titles['seasons'].fillna(0)

# Fill IMDb & TMDB scores with mean (better than dropping)
titles['imdb_score'] = titles['imdb_score'].fillna(titles['imdb_score'].mean())
titles['tmdb_score'] = titles['tmdb_score'].fillna(titles['tmdb_score'].mean())


# 4. Data type corrections

titles['release_year'] = titles['release_year'].astype(int, errors='ignore')
titles['seasons'] = titles['seasons'].astype(int, errors='ignore')

# 5. Clean text columns

titles['title'] = titles['title'].str.strip()
titles['genres'] = titles['genres'].str.strip()
titles['production_countries'] = titles['production_countries'].str.strip()

# 6. (Optional) Merge datasets

merged_df = pd.merge(titles, credits, on="id", how="left")
print("\nMerged dataset shape:", merged_df.shape)


````

### What all manipulations have you done and insights you found?

Answer Here.

**Manipulations Done:**

Removed duplicate rows from both datasets (3 from Titles and 56 from Credits).

**Handled missing values:**

Filled age_certification with "Unknown".

Filled seasons and runtime with 0.

Replaced missing imdb_score and tmdb_score with their mean values.

Filled missing character_name in Credits with "Unknown".

Converted data types for release_year and seasons to integers.

Cleaned text columns (title, genres, production_countries) by removing extra spaces.

Merged the Titles and Credits datasets on id to create a combined dataset with 19 columns and ~125k rows.

**Insights Found:**

The dataset contained only a few duplicates which were removed.

A significant amount of missing data was present in age_certification and seasons, which required careful handling.

seasons is only relevant for TV shows, while movies always have 0 seasons.

Ratings (imdb_score, tmdb_score) had missing values but filling them with mean preserved overall consistency.

After merging, we now have a comprehensive dataset that includes both content information and cast/directors, making it ready for deeper analysis.

Runtime and scores may show weak or no correlation, indicating that length does not strongly affect ratings.
